```python
import pandas as pd
```

```python
df=pd.read_csv("d:\\DataAnalytics\\DataSets\\netflix_titles.csv")
```

```python
import datetime
```

```python
df['director']=df['director'].fillna("No Director Listed")
df['cast']=df['cast'].fillna('No cast listed')
df['country']=df['country'].fillna('Unknown')

```

```python
df['date_added'] = pd.to_datetime(df['date_added'], errors='coerce')
```

```python
df['date_added'] = pd.to_datetime(df['date_added'], format='%B %d, %Y')
```

```python
df['date_added'] = pd.to_datetime(df['date_added'])
```

```python
df['year_added'] = df['date_added'].dt.year
df['month_added'] = df['date_added'].dt.month
```

```python
import matplotlib as plt
```

```python
import seaborn as sns
```

```python
plt.figure(figsize=(5,3))
sns.countplot(x='year_added', hue='type', data=df)
plt.title("Netflix Content Added Over Time")
plt.show()
```

```python
import matplotlib as plt
top_countries = df['country'].value_counts().head(10)
sns.barplot(x=top_countries.values, y=top_countries.index)
plt.title("Top 10 Countries by Netflix Content")
plt.show()
```
